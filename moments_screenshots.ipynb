```python
import pandas
import pandasql
```

```python
moments_screenshots_df = pandas.read_csv('~/Downloads/moments_screenshots.csv')
app_info_df = pandas.read_csv('~/Downloads/app_info.csv')
```

```python
pysqldf = lambda q: pandasql.sqldf(q, globals())
```

```python
moments_screenshots_df
```

```python
query = """
    SELECT 
          app_info_df.org
        , app_info_df.app_name
        , moments_screenshots_df.app_id
        , moments_screenshots_df.last_moment_date
    FROM moments_screenshots_df
    LEFT JOIN app_info_df
        ON app_info_df.app_id = moments_screenshots_df.app_id
    ORDER BY   
          app_info_df.org
        , app_info_df.app_name
        , moments_screenshots_df.app_id
"""
customers_with_screenshots = pysqldf(q=query)
```

```python
customers_with_screenshots
```

```python
customers_with_screenshots.to_csv('~/Downloads/customers with screenshots(moments).csv', sep='|', index=False)
```
