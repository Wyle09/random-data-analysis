# Moments Duration

### Query: 
```sql
SELECT
    user_id,
    UPPER(REPLACE(toString(device_id), '-', '')) AS embrace_id,
    session_id,
    moment_id,
    moment_name,
    date,
    time AS start_time,
    raw_offset,
    duration
FROM moment
WHERE app_id = 'FxdUD'
AND date >= date_add(DAY, -3, now())
ORDER BY
    user_id ASC,
    time DESC
INTO OUTFILE 'FxdUD/FxdUD_moments_last_3_days.csv'
FORMAT CSVWithNames
;
```

````python
import pandas
````

````python
moments = pandas.read_csv('~/Documents/data-analysis/data-sources/FxdUD_moments_last_3_days.csv')
moments
````

### Udpate Objects data types to pandas datatypes

> Time is the moment start time, raw_offset is the millis (CH can’t store milli granularity for ts) and the duration gives you when it ends And the duration is in millis

````python
moments['start_time'] = pandas.to_datetime(moments['start_time'])
moments[['raw_offset', 'duration']] = moments[['raw_offset', 'duration']].apply(pandas.to_timedelta)
moments.dtypes
````

````python
moments['start_time2'] = moments['start_time'] + moments['raw_offset']
moments
````

````python
date_format_string = "%d/%m/%y %H:%M"
moments['end_time'] = pandas.to_datetime((moments['start_time2'] + moments['duration']), format=date_format_string)
moments
````

````python
moments['end_time2'] = pandas.to_datetime(moments['end_time'], format='%Y-%m-%d')
moments
````
