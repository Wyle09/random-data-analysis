```python
import pandas
```

### Load & Transform File

```python
raw_df = pandas.read_csv('data-sources/embrace_data.csv')
raw_df
```

```python
raw_df.dtypes
```

```python
raw_df[['ts_datetime', 'received_ts_datetime']] = raw_df[['ts', 'received_ts']].apply(pandas.to_datetime, unit='s')
raw_df
```

### Filtering Parameters

```python
app_version = 6.1
app_id = 'akCh6'
start_date, end_date = {'2022-05-23 00:00:00', '2022-05-23 23:59:59'}
start_date = pandas.to_datetime(start_date)
end_date = pandas.to_datetime(end_date)
```

```python
df2 = raw_df[
    ((raw_df['received_ts_datetime'] >= start_date) & (raw_df['received_ts_datetime'] <= end_date))
    & ((raw_df['ts_datetime'] >= start_date) & (raw_df['ts_datetime'] <= end_date))
    & (raw_df['app_version'] == app_version)
    & (raw_df['app_id'] == app_id)
]
df2
```
